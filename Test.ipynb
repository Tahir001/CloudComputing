```python
def isPrime(n):
    # Write your code here
    result = 0
    divisible_lst = []
    for i in range(2, (n//2) +1):
        if (n % i == 0):
            result +=1
            divisible_lst.append(i)
    answer = 0
    if result == 0:
        answer = 1
    else:
        divisible_lst.sort()
        answer = divisible_lst[0]
    return answer
    
```
